```python
print(aus_weather['State'].unique())
```

# Victoria Weather Data Preprocessing for Machine Learning

```python
victoria_weather=aus_weather[aus_weather['State']=='VIC']
victoria_weather.head()
```

```python
victoria_weather_filtered=victoria_weather[['Station','Year','Month','Date','Min °C','Max °C','Rain(mm)',
                                            'RH -% 9:00AM','RH -% 3:00PM','Spd - 9:00AM - km/h','Spd - 3:00PM - km/h']]
```

```python
victoria_weather_filtered.head()
```

```python
# Calculate average temperature
victoria_weather_filtered['avg_temp'] = (victoria_weather_filtered['Min °C'] + 
                                         victoria_weather_filtered['Max °C']) / 2

# Calculate average RH
victoria_weather_filtered['avg_rh'] = (victoria_weather_filtered['RH -% 9:00AM'] + 
                                       victoria_weather_filtered['RH -% 3:00PM']) / 2

# Calculate average wind speed
victoria_weather_filtered['avg_windspeed'] = (victoria_weather_filtered['Spd - 9:00AM - km/h'] + 
                                              victoria_weather_filtered['Spd - 3:00PM - km/h']) / 2
```

```python
# Selecting specific columns from the original DataFrame
victoria_weather_filtered_2 = victoria_weather_filtered[['Station', 'Year', 'Month', 'Date', 'Rain(mm)', 'avg_temp', 'avg_rh', 'avg_windspeed']]
```

```python
victoria_weather_filtered_2.isnull().sum()
```

```python
# Identify columns from which to compute the mean (excluding 'Year', 'Month', and 'Date')
columns_to_fill = victoria_weather_filtered_2.columns.difference(['Station','Year', 'Month', 'Date'])

# Calculate the mean for these columns
mean_values = victoria_weather_filtered_2[columns_to_fill].mean()

# Fill missing values in these columns with their respective means
victoria_weather_filtered_2[columns_to_fill] = victoria_weather_filtered_2[columns_to_fill].fillna(mean_values)
```

```python
victoria_weather_filtered_2.isnull().sum()
```

```python
# Define a dictionary to map month names to their numeric representations
month_mapping = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
    'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12
}

# Convert month names to numeric representations
victoria_weather_filtered_2['Month'] = victoria_weather_filtered_2['Month'].map(month_mapping)

victoria_weather_filtered_2['Month'].unique()
victoria_weather_filtered_2['Date'] = victoria_weather_filtered_2['Date'].astype(int)

# Define a default day value, such as 1
default_day = 1

# Create a new date column by combining "Year", "Month", and "Date"
victoria_weather_filtered_2['date'] = pd.to_datetime(victoria_weather_filtered_2['Year'].astype(str) + 
                                                     '-' + victoria_weather_filtered_2['Month'].astype(str) +
                                                     '-' + victoria_weather_filtered_2['Date'].astype(str), 
                                                     errors='coerce')
```

```python
victoria_weather_filtered_2=victoria_weather_filtered_2.rename(columns={'Station':'station'})

victoria_weather_filtered_2=victoria_weather_filtered_2.rename(columns={'Rain(mm)':'rainfall'})
```

```python
train_data_victoria_weather=victoria_weather_filtered_2[['station','date','rainfall','avg_temp','avg_rh','avg_windspeed']]
```

```python
train_data_victoria_weather.to_csv('train_data_victoria_weather')
```
